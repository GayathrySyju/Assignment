```python
import pandas as pd
dataset1 = {'ID':[101,102,103,104],'Name':['Ammu','Binu','Cina','Jinu']}
a1 = pd.DataFrame(dataset1)
dataset2 = {'ID':[102,103,104],'Age':[34,37,45]}
a2 = pd.DataFrame(dataset2)
join2 = pd.merge(a1,a2,on='ID',how='inner')
print(join2)
```

```python
join21 = pd.merge(a1,a2,on='ID',how='outer')
print(join21)
```

```python
join22 = pd.merge(a1,a2,on='ID',how='left')
print(join22)
```

```python
join23 = pd.merge(a1,a2,on='ID',how='right')
print(join23)
```
